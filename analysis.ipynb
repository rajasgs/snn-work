```python
from faker import Faker

fake = Faker()

def get_name(delim = ","):
    
    f_name = fake.first_name().upper()
    l_name = fake.last_name().upper()
    
    return f_name + delim + l_name

def get_address():
    
    return fake.street_address().upper()

def get_city():
    
    return fake.city().upper()

def get_state():
    
    return fake.state().upper()

def get_zip():
    
    return fake.postcode().upper()

def get_country():
    
    return fake.country().upper()

# https://faker.readthedocs.io/en/master/providers/faker.providers.address.html
```

```python
for _item in range(2000):
    
    _name = get_name()
    _address = get_address()
    _city = get_city()
    _state = get_state()
    _zip = get_zip()
    _country = 'USA'

    row_dict = {
        'address_name_LEFT' : _name,
        'address_1_LEFT' : _address,
        'address_2_LEFT' : 'NaN',
        'city_LEFT' : _city,
        'state_LEFT' : _state,
        'ZIP_LEFT' : _zip,
        'Country_LEFT' : _country,
        
        'address_name_RIGHT' : _name,
        'address_1_RIGHT' : _address,
        'address_2_RIGHT' : '',
        'city_RIGHT' : _city,
        'state_RIGHT' : _state,
        'ZIP_RIGHT' : _zip,
        'Country' : _country,
        'Result' : 1,
    }
    
    # print(row_dict)

    df = df.append(row_dict, ignore_index = True)
```
